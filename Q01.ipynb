```python
# imports
import pandas as pd
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

# Data Preprocessing
data = pd.read_csv('./Crop_Dataset.csv')

X = data[['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']]

le = LabelEncoder()
data['crop_label'] = le.fit_transform(data['Label'])

y = data['crop_label']

scaler = StandardScaler()
scaled_data_X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(scaled_data_X, y, test_size=0.2, random_state=42)

# Model Training - Using Support Vector Machine (SVM)
model = SVC(kernel='rbf', probability=True, random_state=42)  # You can choose different kernels like 'linear', 'poly', etc.
model.fit(X_train, y_train)

# Model Evaluation
predicted_y = model.predict(X_test)
predicted_crops = le.inverse_transform(predicted_y)

print(predicted_crops)

# Get the probabilities for each class (crop) predicted by the model
predicted_probability_y = model.predict_proba(X_test)
predicted_crops_prob = le.inverse_transform(predicted_probability_y)

# Get the top 3 crops with the highest predicted probabilities for each sample in the testing data
top_3_crops = predicted_crops_prob.argsort()[:, -3:][:, ::-1]

X_original = scaler.inverse_transform(X_test)

for i, crops in enumerate(top_3_crops):
    print(f"{X_original[i][0]}, {X_original[i][1]}, {X_original[i][2]} Sample {i+1}: Top 3 Crops - {crops}\n")

# Prediction
N = float(input("Enter N: "))
P = float(input("Enter P: "))
K = float(input("Enter K: "))
temperature = float(input("Enter temperature: "))
humidity = float(input("Enter humidity: "))
ph = float(input("Enter ph: "))
rainfall = float(input("Enter rainfall: "))

user_input = pd.DataFrame({
    'N': [N],
    'P': [P],
    'K': [K],
    'temperature': [temperature],
    'humidity': [humidity],
    'ph': [ph],
    'rainfall': [rainfall],
})

scaled_test_data = scaler.fit_transform(user_input)

prediction = model.predict(scaled_test_data)
print(prediction)

probability_prediction = model.predict_proba(scaled_test_data)

top_3_crops2 = probability_prediction.argsort()[:, -3:][:, ::-1]

original_test_prediction = scaler.inverse_transform(scaled_test_data)
print(top_3_crops2)
for i, crops in enumerate(top_3_crops2):
    print(f"Sample {i+1}: Top 3 Crops2 - {crops}\n")
```
